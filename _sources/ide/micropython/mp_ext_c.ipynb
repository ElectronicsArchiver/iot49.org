# External C modules

MicroPython can be extended with [external C modules](https://docs.micropython.org/en/latest/develop/cmodules.html#).

## Unix port

```python
%%service arm32

cd ~/micropython/ports/unix
make clean 
make submodules
make USER_C_MODULES=../../examples/usercmodule
```

```python
%%bash

cd ~/micropython/ports/unix
./micropython -c "import cexample; print(cexample.add_ints(4, -7))"
```

## stm32 port

```python
%%service arm32

cd ~/micropython/ports/stm32
# make clean 
# make submodules
make USER_C_MODULES=../../examples/usercmodule/micropython.cmake
```

## esp32 port

Place custom modules in `~/micropython_modules`.

```python
%%service esp-idf

cd ~/micropython/ports/esp32
# make submodules
# make clean 
make BOARD=GENERIC_OTA \
     USER_C_MODULES=../../../../micropython_modules/micropython.cmake \
     deploy
```

```python
%connect esp32
import cexample
print(cexample.add_ints(5, 7))
```

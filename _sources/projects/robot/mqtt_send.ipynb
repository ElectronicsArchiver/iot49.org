```python
import paho.mqtt.client as mqtt
from time import sleep
import json, socket

TOPIC_ROOT = "vis"

def my_ip():
    with socket.socket(socket.AF_INET, socket.SOCK_DGRAM) as s:
        s.connect(("8.8.8.8", 80))
        return s.getsockname()[0]
    
client = mqtt.Client(clean_session=True, protocol=mqtt.MQTTv311)
print("connect ...")
client.connect(my_ip(), port=1883, keepalive=60)


for i in range(100):
    print(f"publish {TOPIC_ROOT}/ds/lines/new")
    client.publish(f"{TOPIC_ROOT}/ds/lines/new", json.dumps({ "columns": [ "x", "y1", "y2" ] }))
    sleep(0.1)

for i in range(100):
    print(f"publish {TOPIC_ROOT}/ds/lines/add")
    client.publish(f"{TOPIC_ROOT}/ds/lines/add", 
                   json.dumps({ "x": i, "y1": 2*i, "y2": i*i }))
    sleep(1)
```
